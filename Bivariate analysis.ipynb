```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.barplot(x='class', y='age', data=df)
```

```python
sns.barplot(x='class', y='age', hue='sex', data=df)
```

```python
sns.boxplot(x='class', y='age', data=df)
```

```python
sns.boxplot(x='class', y='age', hue='sex', data=df)
```

```python
# violinplot give both boxplot and probability distribution
sns.violinplot(x='class', y='age', data=df)
```

```python
sns.distplot(df[df['class']=='Second']['age'])
```

# Categorical

```python
sns.countplot(data=df, x='survived', hue='sex')
```

```python
x = pd.crosstab(df['survived'], df['class'])
```

```python
x
```

```python
sns.heatmap(x, annot=True, fmt='.1f', cmap='coolwarm')
```

```python
sns.barplot(data=df, x='class', y='age', hue='sex')
```

```python
# to see how the data is distributed
sns.stripplot(x='class', y='age', hue='sex', data=df)
```
