### Q1. Find Top 5 apps which has more installs and rating compare to other from same category I want?

```python
def top_apps(df, sorting_list=['Rating']):
    # first of all create an empty list, which will store result we want
    apps = {}
    # cateogries we have
    categories = df['Category'].unique()
    for cat in categories:
        new_df = df[df['Category'] == cat]
        new_df.sort_values(by=sorting_list, ascending=False, inplace=True)
        apps[cat] = new_df.head(5)
    return apps
    
```

```python
apps = top_apps(df, ['Installs', 'Rating'])
apps.keys()
```
